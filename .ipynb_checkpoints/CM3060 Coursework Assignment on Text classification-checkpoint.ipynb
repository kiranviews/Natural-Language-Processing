## Evaluation methodology (490 words)

To conduct the sentiment analysis, I have applied 3 models on the Coursera review dataset after pre-processing it. I will be applying **Naive Bayes (using CountVectoriszer)**, **Logistic Regression (using TF-IDF)** and **Random Forest classifier (with TF-IDF)** classification models on the cleaned, processed data. 

**Extraction techniques:**

**CountVectorizer**: Converts a collection of text into matrix token counts which is then used as inputs for NLP. As this method implements the bag-of-words model, which represents a document as an unordered collection of words, ignoring grammar and word order but considering word frequencies. This makes this model a good baseline for this text classification project.

**TF-IDF**: This is a popular feature extraction technique as it represents the importance of a term in a document relative to a collection of  documents. TF-ID is used to represent the importance of worlds in a document and it enables effective feature discrimination and aiding in noise reduction for this task.

**Classifiers and Models:** 

**Naive Bayes:** This is a simple probabilistic classifier that applies the Bayes’ theorem with the assumption of feature independence, making it efficient and well-suited for this project. Due to its simplicity, fast training and inference times and competitive performance, I have chosen this as my baseline model. 

**Logistic Regression:** This is a statistical model used for binary classification that can be used to predict the probability of a label belonging to a particular review in context to this project. Since it is widely used due to its simplicity and interpretability, I have chosen this as one of my models.

**Random Forest Classifier:** This model can be utilised to effectively classify text documents into positive, neutral and negative sentiment categories. The model’s ability to handle non-linear relationships and capture feature importance makes it suitable for this text classification project as it is expected to produce reliable and interpretable sentiment predictions.

Upon classification, I will be generating evaluation metrics reports with precision, recall, f-score and accuracy in it. 

**Precision = true positives / (true positives + false positives)**
Precision indicates how reliable the model is in correctly identifying positive sentiments. It measures the proportion of true positive predictions among all positive predictions made by the model.

**Recall = true positives / (true positives + false negatives)**
Recall indicates how well the model captures positive sentiments. It measures the proportion of true positive predictions among all actual positive instances of the data.

**F-score = (2 x Recall x Precision) / (Recall + Precision)**
F-score is the harmonic mean of precision and recall.

**Accuracy = Number of correct predictions / Total number of predictions**
Accuracy measures the overall correctness of the model’s predictions. In doing so, it provides an overall view of the model’s performance. 


By generating a metric report containing the above mentioned metrics, I will have a comprehensive evaluation of the models. Precision, recall, and F1-score will provide me with  insights into the models' performance in identifying positive sentiment accurately. Accuracy will give me an overall measure of model correctness. By doing so, I can compare the performance of both the models and evaluate a better suited model for this project.

## Cleaning the dataset

In this section, I will be cleaning the dataset by removing the hashtags, mentions (@) and URLs and punctuations. Following that, I will be standardising the text to all lowercase. Next, I will be removing all the stopwords. Finally I will be performing stemming

```python
#remove hashtags
review_df['Review'].replace( { r"#(\w+)" : '' }, inplace= True, regex = True)
review_df.head()
```

```python
#Remove Mention "@"
review_df['Review'].replace( { r"@(\w+)" : '' }, inplace= True, regex = True)

review_df.head()
```

```python
#Remove URL
review_df['Review'].astype(str).replace( { r"http\S+" : '' }, inplace= True, regex = True)

review_df.head()
```

```python
#punctuation removal
import string

# Define a function to remove punctuations
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

# Apply the remove_punctuation function to the 'Review' column
review_df['Review'] = review_df['Review'].replace({ r"[{}]".format(string.punctuation) : '' }, regex=True)

review_df.head()
```

```python
#make it all lowercase
review_df['Review']=review_df['Review'].str.lower()

review_df.head()
```

```python
# Download stopwords data
nltk.download('stopwords')

# Get the set of stopwords
stopword_set = set(stopwords.words('english'))

# Function to remove stopwords from a sentence
def remove_stopwords(sentence):
    tokenized_sentence = nltk.word_tokenize(sentence)
    filtered_sentence = [word for word in tokenized_sentence if word.lower() not in stopword_set]
    filtered_text = ' '.join(filtered_sentence)
    return filtered_text

# Apply stopword removal to the 'Review' column in review_df
review_df['Review'] = review_df['Review'].apply(remove_stopwords)

review_df.head()
```

```python
# Function to perform stemming on a text
def stem_text(text):
    stemmer = PorterStemmer()
    stemmed_words = []
    words = text.split()

    for word in words:
        stemmed_word = stemmer.stem(word)
        stemmed_words.append(stemmed_word)

    stemmed_text = ' '.join(stemmed_words)
    return stemmed_text

# Apply stemming to the 'Review' column
review_df['Review'] = review_df['Review'].apply(stem_text)

# Print the updated DataFrame
review_df.head()
```

# Baseline performance and classification approach

```python
#importing modules related to the models
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import csv
import re 
import string
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

#testing with models
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
# from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
```

```python
y = review_df['Label']
x = review_df['Review']

# Splitting the dataset into training and testing parts
X_temp, X_test, y_temp, y_test = train_test_split(x, y, test_size=0.5, random_state=0)
```

## Baseline model: Naive Bayes using CountVectorizer

Naive Bayes will be the basline model used for this sentiment analysis. I will be using the CountVectorizer for this baseline

```python
# Creating CountVectorizer
vectorizer_count = CountVectorizer(max_features=1000)
X_temp_counts = vectorizer_count.fit_transform(X_temp)
X_test_counts = vectorizer_count.transform(X_test)
```

```python
# Naive Bayes with CountVectorizer
classifier_NB_counts = MultinomialNB()
classifier_NB_counts.fit(X_temp_counts, y_temp)
y_predict_NB_counts = classifier_NB_counts.predict(X_test_counts)
accuracy_NB_counts = classifier_NB_counts.score(X_test_counts, y_test)

metrics_NB_counts = metrics.classification_report(y_test, y_predict_NB_counts, output_dict=True)

print("Naive Bayes (CountVectorizer):")
print(metrics.confusion_matrix(y_test, y_predict_NB_counts))
print(metrics.classification_report(y_test, y_predict_NB_counts))
print("Accuracy:", accuracy_NB_counts)
print()
```

## Logistic regression using TF-IDF

```python
# Creating TfidfVectorizer
vectorizer_tfidf = TfidfVectorizer(max_features=1000)
X_temp_tfidf = vectorizer_tfidf.fit_transform(X_temp)
X_test_tfidf = vectorizer_tfidf.transform(X_test)
```

```python
# Logistic Regression with TF-IDF
regressor_LR_tf = LogisticRegression(C=1.0, penalty='l2', solver='newton-cg')
regressor_LR_tf.fit(X_temp_tfidf, y_temp)
y_predict_LR_tf = regressor_LR_tf.predict(X_test_tfidf)
accuracy_LR_tf = regressor_LR_tf.score(X_test_tfidf, y_test)
metrics_LR_tf = metrics.classification_report(y_test, y_predict_LR_tf, output_dict=True)

print("Logistic Regression (TF-IDF):")
print(metrics.confusion_matrix(y_test, y_predict_LR_tf))
print(metrics.classification_report(y_test, y_predict_LR_tf))
print("Accuracy:", accuracy_LR_tf)
print()
```

## Random Forest classifier with TF-IDF

```python
#Random Forest classifier with TF-IDF
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_temp_tfidf, y_temp)
y_predict_rf = random_forest.predict(X_test_tfidf)
accuracy_rf = random_forest.score(X_test_tfidf, y_test)
metrics_rf = metrics.classification_report(y_test, y_predict_rf, output_dict=True)

print("Random Forest (TF-IDF):")
print(metrics.confusion_matrix(y_test, y_predict_rf))
print(metrics.classification_report(y_test, y_predict_rf))
print("Accuracy:", accuracy_rf)
print()
```

# Evaluation

## Compilation of all metrics

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary to store the metrics
model_metrics = {
    'Model': ['Naive Bayes (CountVectorizer)', 'Logistic Regression (TF-IDF)', 'Random Forest (TF-IDF)'],
    'Accuracy': [accuracy_NB_counts, accuracy_LR_tf, accuracy_rf],
    'Precision': [metrics_NB_counts['weighted avg']['precision'], metrics_LR_tf['weighted avg']['precision'], metrics_rf['weighted avg']['precision']],
    'Recall': [metrics_NB_counts['weighted avg']['recall'], metrics_LR_tf['weighted avg']['recall'], metrics_rf['weighted avg']['recall']],
    'F1-score': [metrics_NB_counts['weighted avg']['f1-score'], metrics_LR_tf['weighted avg']['f1-score'], metrics_rf['weighted avg']['f1-score']]
}

# Create a DataFrame to compare the metrics of all models
df = pd.DataFrame(model_metrics)

# Print the comparison table
print("Comparison of Model Metrics:")
print(df)
print()

```

## Plotting the metrics for evaluation

```python
# Plot the metrics
plt.figure(figsize=(15, 6))
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1-score']
for i, metric in enumerate(metrics_names):
    plt.subplot(2, 2, i+1)
    plt.bar(df['Model'], df[metric])
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.ylim(0.65,0.80)
    plt.title(f'Comparison of {metric} for Models')

plt.tight_layout()
plt.show()
```
